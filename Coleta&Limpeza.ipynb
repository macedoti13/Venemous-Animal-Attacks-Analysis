# Dados da SINAN sobre Ataques de Animais Peçonhentos

```python
import pandas as pd
import numpy as np
import warnings
from pysus.online_data import SINAN
from pysus.preprocessing.decoders import decodifica_idade_SINAN

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore') # eu sei que isso é feio, mas era necessário 
```

## Coleta Utilizando a lib PySUS

```python
df = SINAN.download(year=2017, disease='Animais Peçonhentos', state='')
display(df.head())
df.info()
```

## Limpeza de Dados

**Transformando em um DataFrame do Pandas**

```python
df = pd.DataFrame(df)
```

**Decodificando a Idade**

```python
df['idade'] = decodifica_idade_SINAN(df.NU_IDADE_N, 'Y')
df[['NU_IDADE_N', 'idade']].head()
```

**Removendo colunas desnecessárias**

```python
df.drop(columns=['ANT_LOCALI', 'SG_UF_NOT','ID_MUNICIP', 'NU_IDADE_N', 'DT_NASC', 'TP_NOT', 'ID_AGRAVO', 'SEM_NOT', 'NU_ANO', 'ID_REGIONA', 'SEM_PRI', 'CS_GESTANT', 
                 'CS_RACA', 'CS_ESCOL_N', 'SG_UF', 'ID_MN_RESI', 'ID_RG_RESI', 'ID_PAIS', 'DT_INVEST', 'ID_OCUPA_N', 'ANT_TEMPO_', 'CLI_LOCA_1', 'CLI_OUTR_2', 'CLI_OUTR_3', 
                 'CLI_TEMPO_', 'NU_AMPOL_1', 'NU_AMPOL_3', 'NU_AMPOL_4', 'NU_AMPO_5', 'NU_AMPOL_6', 'NU_AMPO_7', 'NU_AMPOL_8', 'NU_AMPOL_9', 'COM_LOC', 'COM_SECUND', 
                 'COM_NECROS', 'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM', 'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE', 'DOENCA_TRA', 'DT_ENCERRA', 'MCLI_LOCAL', 
                 'CLI_LOCAL_', 'MCLI_SIST', 'ANI_TIPO_1', 'DT_DIGITA', 'ANI_SERPEN', 'ANI_ARANHA', 'ANI_LAGART', 'COM_COMPOR', 'DT_OBITO', 'DT_NOTIFIC', 'ANT_MUNIC_'], 
                 axis=1, inplace=True)
df.info()
```

**Renomeando as Colunas para melhor interpretabilidade**

```python
df.rename(columns={'DT_SIN_PRI':'dt_sintomas', 'CS_SEXO':'sexo', 'ANT_DT_ACI':'dt_acidente', 'ANT_UF':'uf', 'ANT_ZONA':'zona_acidente', 'ANT_LOCA_1':'local_picada', 
                   'CLI_DOR':'dor', 'CLI_EDEMA':'edema', 'CLI_EQUIMO':'esquimose', 'CLI_NECROS':'necrose', 'CLI_NEURO':'neuroparalisia', 'CLI_HEMORR': 'hemorragia', 
                   'CLI_VAGAIS':'vomito/diarreia', 'CLI_MIOLIT':'miolise', 'CLI_RENAL':'sintomas_renais', 'TP_ACIDENT':'tp_animal', 'TRA_CLASSI':'envenenamento', 
                   'CON_SOROTE':'soroterapia', 'NU_AMPOLAS':'qtd_ampolas', 'EVOLUCAO':'obito'}, 
                   inplace=True)
df.info()
```

**Alterando o Tipo de Dado das Colunas Para o Tipo Correto**

```python
for col in df.columns:
    if col.startswith('dt_'):
        df[col] = pd.to_datetime(df[col])
    else:
        try:
            df[col] = pd.to_numeric(df[col])
        except ValueError:
            continue

df.info()
```

**Removendo Valores Duplicados**

```python
print('Quantidade de duplicadas:', df.duplicated().sum())
print('Removendo Duplicadas...')
df.drop_duplicates(inplace=True)
print('Quantidade de Duplicadas após Remoção:', df.duplicated().sum())
```

**Tratando os Valores Nulos**

```python
df.fillna(np.nan)
df.isnull().sum()
```

**Removendo as linhas onde não é possível preencher os nulos**

```python
df.dropna(subset=['dt_acidente', 'tp_animal', 'idade', 'zona_acidente'], inplace=True)
df.isnull().sum()
```

```python
df.head(10)
```

**Decodificando Colunas Categóricas**

_Codificação Oficial das Colunas_
- 1: Sim
- 2: Não
- 9: Ignorado

_Método Utilizado para Preenchimento de Nulos e Decodificação dos Valores_
- 1: Recebe 1
- 2, 9 ou Nulo: Recebe 0

```python
cols = ['dor', 'edema', 'esquimose', 'necrose', 'neuroparalisia', 'hemorragia', 'vomito/diarreia', 'miolise', 'sintomas_renais', 'soroterapia']
for col in df.columns:
    if col in cols:
        df[col] = df[col].apply(lambda x: 1 if x == 1 else 0)

df[cols].head()
```

**Codificação Oficial da Coluna Óbito (indica se a pessoa atacada veio a óbito). Coluna Original: EVOLUCAO:**
- 1: Curado
- 2: Óbito devido ao Ataque do Animal (nosso interesse)
- 3: Óbito por outro motivo  
- 9: Ignorado

**Método Utilizado para Preenchimento de Nulos e Decodificação dos Valores**
- 2: Recebe 1 
- 1, 3 ou 9: Recebe 0

```python
df['obito'] = df['obito'].apply(lambda x: 1 if x == 2 else 0)
df['obito'].head()
```

**Decodificações Personalizadas de acordo com o dicionário de dados disponibilizado pelo SINAN**

```python
def tp_animal(x):
    if x == 1: return 'serpente'
    if x == 2: return 'aranha'
    if x == 3: return 'escorpiao'
    if x == 4: return 'lagarta'
    if x == 5: return 'abelha'
    return 'outro'

def envenenamento(x):
    if x == 1: return 'leve'
    if x == 2: return 'moderado'
    if x == 3: return 'grave'
    return 'nenhum'

def local_picada(x):
    if x == 1:  return 'cabeça'
    if x == 2:  return 'braco'
    if x == 3:  return 'antebraco' 
    if x == 4:  return 'mao' 
    if x == 5:  return 'dedo_mao'
    if x == 6:  return 'tronco' 
    if x == 7:  return 'coxa' 
    if x == 8:  return 'perna' 
    if x == 9:  return 'pe'
    if x == 10: return 'dedo_pe' 
    return 'outro' 

def zona_acidente(x):
    if x == 1: return 'urbana' 
    if x == 2: return 'rural' 
    if x == 3: return 'periurbana' 
    return 'outra'

df['tp_animal'] = df['tp_animal'].apply(tp_animal)
df['envenenamento'] = df['envenenamento'].apply(envenenamento)
df['local_picada'] = df['local_picada'].apply(local_picada)
df['zona_acidente'] = df['zona_acidente'].apply(zona_acidente)
df.head()
```

```python
df.isnull().sum()
```

**Trocando o Tipo de Dado das Colunas Inteiras para int64**

```python
df[['uf', 'idade']] = df[['uf', 'idade']].astype('int64')
df.info()
```

**Decodificando a Coluna _uf_ com os Dados Oficiais do IBGE**

```python
df_ibge = pd.read_excel('data/estados.xls')
df_ibge = df_ibge[['UF', 'Nome_UF']]
df_ibge.rename(columns={'Nome_UF':'estado', 'UF':'uf'}, inplace=True)
df_ibge.drop_duplicates(inplace=True)
df_ibge.head()
```

**Juntando os dataframes para trazer a coluna com o nome do estado**

```python
# juntando os dataframes para trazer a coluna com o nome do estado
df = df.merge(df_ibge)

# removendo a coluna com o estado codificado
df.drop(columns=['uf'], axis=1, inplace=True)
```

**Reordenando as Colunas para Melhor Visualização**

```python
df = df[['dt_acidente', 'dt_sintomas', 'sexo', 'idade', 'estado', 'zona_acidente', 'local_picada', 'dor', 'edema', 'esquimose', 'necrose', 'neuroparalisia', 
         'hemorragia', 'vomito/diarreia', 'miolise', 'sintomas_renais', 'tp_animal', 'envenenamento', 'soroterapia', 'qtd_ampolas', 'obito']]

df.head()
```

**Tratando os Outliers**

**Problema**: 
- Registros onde a pessoa passou por soroterapia, porém, o número de ampolas utilizadas consta como 0

**Solução**:
- Dependendo do envenenamento que essa pessoa teve, trocamos a qtd_ampolas para a média de ampolas daquela envenenamento

```python
df.groupby(['envenenamento'])['qtd_ampolas'].mean()
```

```python
df.qtd_ampolas = np.where(((df.qtd_ampolas==0) & (df.soroterapia==1) & (df.envenenamento=='grave')),4, df.qtd_ampolas)
df.qtd_ampolas = np.where(((df.qtd_ampolas==0) & (df.soroterapia==1) & (df.envenenamento=='moderado')),2, df.qtd_ampolas)
df.qtd_ampolas = np.where(((df.qtd_ampolas==0) & (df.soroterapia==1) & (df.envenenamento=='leve')),1, df.qtd_ampolas)
df.qtd_ampolas = np.where(((df.qtd_ampolas==0) & (df.soroterapia==1) & (df.envenenamento=='nenhum')),1, df.qtd_ampolas)
df[(df['qtd_ampolas']==0)&(df['soroterapia']==1)]
```

**Problema**:
- Registros onde os sintomas da picada do animal surgiram antes da própria picada

**Solução**:
- Alteramos a data dos sintomas para a mesma data da picada

```python
df.dt_sintomas = np.where((df.dt_acidente > df.dt_sintomas), df.dt_acidente, df.dt_sintomas)
df[df.dt_acidente > df.dt_sintomas]
```

**Remoção de Outliers**

**Problema**:
- Pessoas que tiveram sintomas da picada muito depois da data da picada

**Solução**:
- Removemos todos os registros onde a data dos sintomas vem mais de 15 dias depois da picada

```python
df.drop(df[(df.dt_sintomas - df.dt_acidente) > '15 days'].index, inplace=True)
```

**Problema**:
- Registros de óbitos em que a pessoa não teve nenhum envenenamento

```python
df.drop(df[(df.obito == 1) & (df.envenenamento=='nenhum')].index, inplace=True)
```

**Problema:**
- Registros de óbitos onde a pessoa não teve nenhum sintoma

```python
df.drop(df[(df.obito==1)&(df.dor==0)&(df.edema==0)&(df.esquimose==0)&(df.necrose==0)&(df.neuroparalisia==0)
        &(df.hemorragia==0)&(df['vomito/diarreia']==0)&(df.miolise==0)&(df.sintomas_renais==0)].index, inplace=True)
```
